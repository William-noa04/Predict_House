```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns
from sklearn import datasets
```

```python
data = pd.read_csv("train.csv")
data.head()
```

```python
data.describe()
```

```python
data.shape
```

```python
data.info()
```

```python
# selection de variables
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
X = data[features]
y = data['SalePrice']
X
y
```

```python
# normalisation
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
y=y.values.reshape(-1,1)
y=scaler.fit_transform(y)
y
```

```python
# donnees d entrainement et de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
# construire l architecture du reseau
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential(
    [
    Dense(64, activation='relu', input_shape=(3,)),
    Dense(32, activation='relu'),
    Dense(1)
    ]
)
```

```python
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mean_absolute_error'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100, validation_split=0.2)
```

```python
# evaluation
model.evaluate(X_test, y_test)
```

```python
# calcul accuracy
test_loss, test_acc = model.evaluate(X_test, y_test)
print('Test accuracy:', test_acc)
print('Test loss',test_loss)
```

```python
features = ['GrLivArea', 'BedroomAbvGr', 'FullBath']
X = data[features]
y = data['SalePrice'].values.reshape(-1, 1)

scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()

X_scaled = scaler_x.fit_transform(X)
y_scaled = scaler_y.fit_transform(y)
```

###  DEVOIR :  meme que le precedent mais avec ce reseau de neurones 

```python
# ====================== FONCTION DE PRГүDICTION (Version corrigГ©e) ======================

def predire_prix_maison(GrLivArea, BedroomAbvGr, FullBath):
    
    nouvelle_maison = np.array([[GrLivArea, BedroomAbvGr, FullBath]])
    
    # === IMPORTANT : Utilise scaler_x et scaler_y ===
    nouvelle_maison_scaled = scaler_x.transform(nouvelle_maison)   # scaler pour X
    
    prediction_scaled = model.predict(nouvelle_maison_scaled, verbose=0)
    
    prix_reel = scaler_y.inverse_transform(prediction_scaled)     # scaler pour y
    
    return round(float(prix_reel[0][0]), 2)


# ====================== TESTS ======================
print("Test 1 :", predire_prix_maison(2000, 3, 2), "$")
print("Test 2 :", predire_prix_maison(1500, 2, 1), "$")
print("Test 3 :", predire_prix_maison(2500, 4, 3), "$")
```
